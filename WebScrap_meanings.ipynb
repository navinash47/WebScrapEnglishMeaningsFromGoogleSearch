```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from tqdm import tqdm
```

```python
def make_url(word):
  url ="https://www.google.com/search?q="+word+"+meaning"
  return url
```

```python
def minisoup(word):
  # headers={"User-Agent": "Mozilla/5.0 (Windows; U; Windows NT 6.1; rv:2.2) Gecko/20110201"}
  response=requests.get(make_url(word))
  
  soup=BeautifulSoup(response.text,'html.parser')
  # print(soup.prettify())
  t_desc=soup.find_all('div',{'class':'kCrYT'})
  if t_desc:
    soup=t_desc[1]
  else: 
    return None
  t_desc=soup.find_all('div',{'class':'Ap5OSd'})
  if t_desc:
    return t_desc
  else:
    return None
```

```python
def dict_find(t_desc):
  i=-1
  word_type=[]
  dict_={}
  if t_desc==None:
    return None
  for t in t_desc:
    dict_each={}
    P=t.find_all(text=True)
    if len(P)==1:
      word_type.append((P[0])[:len(P[0])-1])
      i=i+1
    if len(P)>1:
      meaning=True
      m=[]
      ex=[]
      s=[]
      for p in P:
        if "synonyms: " in p:
          
          s=p[len("synonyms: "):].split(", ")
        else:
          if(meaning):
          
            m.append(p)
            meaning=False
          else:
          
            ex.append(p)
            meaning=True
      dict_each["meanings"]=m
      dict_each["examples"]=ex
      dict_each["synonyms"]=s

      dict_[word_type[i]]=dict_each
  return dict_
```

```python
def find_words(words):
  t_words=[]
  t_keys=[]
  t_meanings=[]
  t_examples=[]
  t_synonyms=[]
  for word in tqdm(words):
  # for word in words:
    # print(word,end="")
    # print()
    each_dict=dict_find(minisoup(word))
    # print(each_dict)
    if(each_dict==None):
      t_examples.append("")
      t_words.append(word)
      t_keys.append("")
      t_meanings.append("")
      t_synonyms.append("")
    else:
      for key in each_dict.keys():
        t_keys.append(key)
        t_words.append(word)
        t_meanings.append(str(each_dict[key]["meanings"]))
        # print(each_dict[key]["synonyms"]),
        if(each_dict[key]["synonyms"]):
          t_synonyms.append(str(each_dict[key]["synonyms"]))
        else:
          t_synonyms.append("")

        if(each_dict[key]["examples"]):
          t_examples.append(str(each_dict[key]["examples"]))
        else:
          t_examples.append("")
  data={"Word":t_words,"Type":t_keys,"Meanings":t_meanings,"Examples":t_examples,"Synonyms":t_synonyms}
  return pd.DataFrame(data)
```
